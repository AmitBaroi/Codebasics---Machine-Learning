```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r'data\1_homeprices.csv')
df
```

```python
model = LinearRegression()
model.fit(df[['area']], df['price'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

# 1. `pickle`

```python
import pickle
```

### Saving trained model using `pickle`

Python's `pickle` module allows us to save Python objects to _.pickle_ files.

```python
# Opeining a file in 'write bytes' mode
with open('model_pickle', 'wb') as file:
    pickle.dump(model, file)
```

### Opening "pickled" model from file

```python
# Opeining a file in 'read bytes' mode
with open('model_pickle', 'rb') as file:
    mp = pickle.load(file)
```

### Using the trained model loaded from file

```python
mp.predict([[5000]])
```

```python
mp.coef_
```

```python
mp.intercept_
```

# 2. `joblib`

`from sklearn.externals import joblib` has been depricated.

`joblib` can now be directly imported, if installed via pip or conda.

```python
import joblib
```

### Saving trained model using `joblib`

Scikit-learn's `joblib` module allows us to save trained models, especially ones with a large number of Numpy arrays much more efficiently.

```python
joblib.dump(model, 'model_joblib')
```

### Opening trained model from `joblib` file

```python
mj = joblib.load('model_joblib')
```

### Using the trained model loaded from file

```python
mj.predict([[5000]])
```

```python
mj.coef_
```

```python
mj.intercept_
```
