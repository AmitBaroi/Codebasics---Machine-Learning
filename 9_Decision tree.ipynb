```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"data\9_data.csv")
df
```

# Preparing data for training model

```python
inputs = df.drop('salary_more_then_100k', axis='columns')
target = df['salary_more_then_100k']
```

### Encoding categorial data for training

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
inputs['company'] = le.fit_transform(inputs['company'])
inputs['job'] = le.fit_transform(inputs['job'])
inputs['degree'] = le.fit_transform(inputs['degree'])
inputs
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(inputs, target)
```

```python
model.score(inputs, target)
```

```python
model.predict([[1, 1, 1]])
```
